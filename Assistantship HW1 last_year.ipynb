Задание 4 - работа с датасетом. 3 балла 

Откройте датафрейм states.csv (убедитесь, что он в той же папке, где ноутбук). С помощью функций .loc или .iloc ответьте на вопросы (не используя гугл):

1) Какие государственные языки в Бельгии?

2) Какое население в Болгарии?

3) Сколько у Польши мест в Европейском парламенте?

```python
states = pd.read_csv("states.csv", index_col = 0)
states 
```

```python
print("Какие государственные языки в Бельгии?")
print(states.loc["Belgium", "Language"])

print("Какое население в Болгарии?")
print(states.loc["Bulgaria", "Population"])

print("Сколько у Польши мест в Европейском парламенте?")
print(states.loc["Poland", "European Parliament Seats"])
```
