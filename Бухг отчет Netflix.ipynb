### Чем отличаются map, apply, lambda?

Ключевое отличие map() и apply() заключается в том, что  map() применяется к спискам, словарям, множествам или кортежам, а apply() - к столбцам (axis = 0) или строкам (axis = 1) датафрейма. lambda применяется к обычным элементам типа числа или строки и тд.  


Встроенная в Python функция map() используется для применения функции к каждому элементу итерируемого объекта (например, **списка, словаря, множества или кортежа**) и возврата нового итератора для получения результатов.Вместо использования цикла for функция map() дает возможность применить функцию к каждому элементу итерируемого объекта. Это повышает производительность, поскольку функция применяется только к одному элементу за раз без создания копий элементов в другом итерируемом объекте.  
Для того, чтобы вернулся список, нужно не забыть обернуть все выражение с map() в **list(), dict(zip())??, set(), tuple()**

```python
# список
l = [1,2,3,4]
l2 = list(map(lambda x: x/2, l))
print(l2)

# словарь
d = {"a":2, "b":7, "c":67}
d2 = list(map(lambda x: x*2, d.values()))
d = dict(zip(d.keys(),d2))# - позволяет создать словарь из двух списков
print(d)

# множество
s = {1,2,3,4}
s = set(map(lambda x: x+1,s))
print(s)

#кортеж
t = (9,5,63)
t = tuple(map(lambda x: x/3,t))
print(t)
```

```python
data = pd.read_excel("C:/Users/Vanessa/Desktop/работа/газпром нефть/Q3'22-Website-Financials.xlsx")
# убрать первые три строки бессодержательные
data = data[3:]

# удалить пустые столбцы
data = data.dropna(how = "all", axis = 1)
data

# сцепить два столбца. Имея NaN не работает склеивание строк, т.к. при наличие хотя бы одного NaN результатом будет NaN
data[['Netflix, Inc.','Unnamed: 1']] = data[['Netflix, Inc.','Unnamed: 1']].fillna("")
data['Netflix, Inc.'] = data['Netflix, Inc.'] + data['Unnamed: 1']
data = data.drop('Unnamed: 1', axis = 1)

# проверка того,что пустых значений не осталось. 
# data.count()

# транспонируем
data = data.T

# заполнить все пропуски вниз (в прямом порядке)
data[3] = data[3].fillna(method = 'ffill')

# удалить индексы строк
data = data.reset_index(drop = True)

# сделать первую строку заголовками
data.columns = data.loc[0]
data  = data.drop(0)

# удалить пустые столбцы
data = data.dropna(how = "all", axis = 1)
data.columns
```

```python
# поменять названия столбцов
data.columns = ['Statement Type', 'Date', 'Year', 'Revenues', 'Cost of revenues', 'Marketing',
       'Technology and development', 'General and administrative',
       'Operating income', 'Interest expense ',
       'Interest and other income (expense)', 'Income before income taxes',
       'Benefit from (provision for) income taxes', 'Net income', 'Earnings per share Basic',
       'Earnings per share Diluted', 'Weighted-average shares of common stock outstanding: Basic', 'Weighted-average shares of common stock outstanding: Diluted']

# сцепить два столбца в один
data["Date"] = data["Date"] + data["Year"].astype(str)
data = data.drop("Year", axis = 1)
data.head()
```

```python
import datetime
# Приведение даты к нужному формату
data['Date'] = pd.to_datetime(data['Date'], format = '%B %d,%Y')
data.head()
```
